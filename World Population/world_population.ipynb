## 🎯 Objective

The goal of this analysis is to examine global country-level data and uncover insights into population, rank of countries, growth rate, countries' population percentages, and other key indicators. We'll answer questions like

- What are the top 10 most populated countries in 2022 and other years?

## 📄 Dataset Description

**Source:** [Kaggle - Countries of the World](https://www.kaggle.com/datasets/iamsouravbanerjee/world-population-dataset)  
**Rows:** 234 countries  
**Columns:** 17 features, including

- `Rank:` Rank by Population.
- `CCA3:` 3 Digit Country/Territories Code.
- `Country/Territories:` Name of the Country/Territories.
- `Capital:` Name of the Capital.
- `Continent:` Name of the Continent.
- `2022 Population:` Population of the country/territories in the year 2022.
- `2020 Population:` Population of the Country/Territories in the year 2020.
- `2015 Population:` Population of the Country/Territories in the year 2015.
- `2010 Population:` Population of the Country/Territories in the year 2010.
- `2000 Population:` Population of the Country/Territories in the year 2000.
- `1990 Population:` Population of the Country/Territories in the year 1990.
- `1980 Population:` Population of the Country/Territories in the year 1980.
- `1970 Population:` Population of the Country/Territories in the year 1970.
- `Area (km²):` Area size of the country/territories in square kilometers.
- `Density (per km²):` Population Density per square kilometer.
- `Growth Rate:` Population Growth Rate by Country/Territories.
- `World Population Percentage:` The population percentage by each country/territory.

Note: Some columns contain missing or inconsistent values that will require cleaning.

## 🧹 Data Cleaning / Preprocessing
- Interpolated missing values across the `population` columns to maintain continuity over time.
- Filled missing values in the `Area (km²)` column using appropriate imputation.
- Calculated and filled missing values in `Density (per km²)` and `Growth Rate` based on population and area data.


```python
# checking for duplicates
df.duplicated().any()
```

```python
# checking for null values
df.isnull().sum().any()
```

```python
# checking columns that contain missing values
df.columns[df.isnull().any()]
```

```python
df.isnull().sum()
```

```python
# rows with missing values in the "2022 Population" column
df[df["2022 Population"].isnull()]
```

```python
# rows with missing values in the dataframe
df[df.isnull().any(axis = 1)]
```

- Because missing values are spread across each column and dropping all missing values would result in losing the entire rows of data of individual countries, we'll fill the missing values with the average across each column.

```python
population_columns = ['2022 Population', '2020 Population', '2015 Population',
       '2010 Population', '2000 Population', '1990 Population',
       '1980 Population', '1970 Population']
```

```python
df[population_columns] = df[population_columns].interpolate(axis = 1, limit_direction = "both")
```

```python
# checking to see the results
df.isnull().sum()
```

```python
df[df.isnull().any(axis = 1)]
```

```python
df.columns
```

```python
df.at[183, "Area (km²)"] = 196722
df.at[227, "Area (km²)"] = 916445
```

```python
for x in df.index:
    if df.loc[x, "Density (per km²)"] == "NaN":
        df.at[x, "Density (per km²)"] = df["2022 Population"]/df["Area (km²)"]
```

```python
# Loop through the index of the DataFrame
for x in df.index:
    # Check if the value in the 'Density (per km²)' column is NaN
    if pd.isna(df.loc[x, "Density (per km²)"]):
        # Calculate the density for that specific row
        df.at[x, "Density (per km²)"] = df.loc[x, "2022 Population"] / df.loc[x, "Area (km²)"]
```

```python
df[df.isnull().any(axis = 1)]
```

```python
# Looping through the index of the dataframe
for x in df.index:
    # check if the values in the 'Growth Rate' column are NaN
    if pd.isna(df.loc[x, "Growth Rate"]):
        # Calculate the growth rate for that specific row
        df.at[x, "Growth Rate"] = (df.loc[x, "2022 Population"] - df.loc[x, "1970 Population"]) / df.loc[x, "1970 Population"]
```

```python
top_pop_2022 = df.groupby("Country").agg({"2022 Population":"sum"}).sort_values(by = "2022 Population", ascending = False).head(10)
top_pop_2022
```

## What countries show the highest increase in growth?

```python
# Option 1
df["population_growth_1970_2022"] = df["2022 Population"] - df["1970 Population"]
```

```python
largest_increasing_pop = df.sort_values(by = "population_growth_1970_2022", ascending = False)
largest_increasing_pop.head(10)
```

```python
# Plot
plt.bar(largest_increasing_pop["Country"].head(10), largest_increasing_pop["population_growth_1970_2022"].head(10), color = "green")

plt.title("Top 10 Countries with Largest Population Increase (1970 - 2022)", fontsize = 16)
plt.xlabel("Country", fontsize = 12)
plt.ylabel("Population Increase", fontsize = 12)
plt.xticks(rotation = 45)

plt.tight_layout()
plt.show()
```

## What are Countries are experiencing a decline in growth?

```python
df["Growth_Rate_1970_2022"] = ((df["2022 Population"] - df["1970 Population"])/df["1970 Population"]) * 100
```

```python
declining_growth_rate = df[df["Growth_Rate_1970_2022"] < 0]
declining_growth_rate_group = declining_growth_rate.groupby("Country").agg({"Growth_Rate_1970_2022":"sum"})
declining_growth_rate_group
```

```python
# Plot
declining_growth_rate_group.plot(kind = "bar")

plt.title("Countries with Declining Population (1970 - 2022)", fontsize = 16)
plt.xlabel("Country", fontsize = 12)
plt.ylabel("Growth Rate (%)", fontsize = 12)

plt.tight_layout()
plt.show()
```
