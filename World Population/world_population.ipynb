## ðŸ§¹ Data Cleaning / Preprocessing
- Interpolated missing values across the `population` columns to maintain continuity over time.
- Filled missing values in the `Area (kmÂ²)` column using appropriate imputation.
- Calculated and filled missing values in `Density (per kmÂ²)` and `Growth Rate` based on population and area data.


```python
# checking for duplicates
df.duplicated().any()
```

```python
# checking for null values
df.isnull().sum().any()
```

```python
# checking columns that contain missing values
df.columns[df.isnull().any()]
```

```python
df.isnull().sum()
```

```python
# rows with missing values in the "2022 Population" column
df[df["2022 Population"].isnull()]
```

```python
# rows with missing values in the dataframe
df[df.isnull().any(axis = 1)]
```

- Because missing values are spread across each column and dropping all missing values would result in losing the entire rows of data of individual countries, we'll fill the missing values with the average across each column.

```python
population_columns = ['2022 Population', '2020 Population', '2015 Population',
       '2010 Population', '2000 Population', '1990 Population',
       '1980 Population', '1970 Population']
```

```python
df[population_columns] = df[population_columns].interpolate(axis = 1, limit_direction = "both")
```

```python
# checking to see the results
df.isnull().sum()
```

```python
df[df.isnull().any(axis = 1)]
```

```python
df.columns
```

```python
df.at[183, "Area (kmÂ²)"] = 196722
df.at[227, "Area (kmÂ²)"] = 916445
```

```python
for x in df.index:
    if df.loc[x, "Density (per kmÂ²)"] == "NaN":
        df.at[x, "Density (per kmÂ²)"] = df["2022 Population"]/df["Area (kmÂ²)"]
```

```python
# Loop through the index of the DataFrame
for x in df.index:
    # Check if the value in the 'Density (per kmÂ²)' column is NaN
    if pd.isna(df.loc[x, "Density (per kmÂ²)"]):
        # Calculate the density for that specific row
        df.at[x, "Density (per kmÂ²)"] = df.loc[x, "2022 Population"] / df.loc[x, "Area (kmÂ²)"]
```

```python
df[df.isnull().any(axis = 1)]
```

```python
# Looping through the index of the dataframe
for x in df.index:
    # check if the values in the 'Growth Rate' column are NaN
    if pd.isna(df.loc[x, "Growth Rate"]):
        # Calculate the growth rate for that specific row
        df.at[x, "Growth Rate"] = (df.loc[x, "2022 Population"] - df.loc[x, "1970 Population"]) / df.loc[x, "1970 Population"]
```

## ðŸ“Š Exploratory Data Analysis (EDA)

### What are the top 10 most populated countries in 2022?

```python
top_pop_2022 = df.groupby("Country").agg({"2022 Population":"sum"}).sort_values(by = "2022 Population", ascending = False).head(10)
top_pop_2022
```

```python
# Plot
plt.figure(figsize = (12, 6))
top_pop_2022.plot(kind = "bar", color = "skyblue")

# Add labels and title
plt.title("Top 10 Most Populated Countries (2022)", fontsize = 16)
plt.xlabel("Country", fontsize = 12)
plt.ylabel("2022 Population", fontsize = 12)


plt.tight_layout()
plt.show()
```
