- Because missing values are spread across each column and dropping all missing values would result in losing the entire rows of data of individual countries, we'll fill the missing values with the average across each column.

```python
population_columns = ['2022 Population', '2020 Population', '2015 Population',
       '2010 Population', '2000 Population', '1990 Population',
       '1980 Population', '1970 Population']
```

```python
df[population_columns] = df[population_columns].interpolate(axis = 1, limit_direction = "both")
```

```python
# checking to see the results
df.isnull().sum()
```

```python
df.at[183, "Area (kmÂ²)"] = 196722
df.at[227, "Area (kmÂ²)"] = 916445
```

```python
for x in df.index:
    if df.loc[x, "Density (per kmÂ²)"] == "NaN":
        df.at[x, "Density (per kmÂ²)"] = df["2022 Population"]/df["Area (kmÂ²)"]
```

```python
# Loop through the index of the DataFrame
for x in df.index:
    # Check if the value in the 'Density (per kmÂ²)' column is NaN
    if pd.isna(df.loc[x, "Density (per kmÂ²)"]):
        # Calculate the density for that specific row
        df.at[x, "Density (per kmÂ²)"] = df.loc[x, "2022 Population"] / df.loc[x, "Area (kmÂ²)"]
```

```python
df[df.isnull().any(axis = 1)]
```

```python
# Looping through the index of the dataframe
for x in df.index:
    # check if the values in the 'Growth Rate' column are NaN
    if pd.isna(df.loc[x, "Growth Rate"]):
        # Calculate the growth rate for that specific row
        df.at[x, "Growth Rate"] = (df.loc[x, "2022 Population"] - df.loc[x, "1970 Population"]) / df.loc[x, "1970 Population"]
```
